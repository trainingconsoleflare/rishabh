## What was the best month for sales? How much was earned that month?

```python
ndf = df.groupby(['Month','Month Name']).agg(
    revenue = ('Amount','sum')
)

ndf = ndf.reset_index() 
ndf.to_csv('Monthly Sales.csv',index=False)
```

## Line Chart

```python
plt.figure(figsize=(10,5))
plt.plot(ndf['Month Name'],ndf['revenue'],marker='o',label='Flipkart Sales',color='Skyblue')
plt.title('Flipkart Sales')
plt.grid() 
plt.legend()
plt.xlabel('Month')
plt.ylabel('Sale')
plt.xticks(rotation=45)
plt.show()
```

## Bar chart

```python
plt.figure(figsize=(10,5))
plt.bar(ndf['Month Name'],ndf['revenue'],label='Flipkart Sales',color='Skyblue')
plt.title('Flipkart Sales')
plt.grid() 
plt.legend()
plt.xlabel('Month')
plt.ylabel('Sale')
plt.xticks(rotation=45)
plt.show()
```

## Total Sale Average Sale

```python
ndf = df.groupby(['Month','Month Name']).agg(
    revenue = ('Amount','sum'),
    avg_sale = ('Amount','mean')
)

ndf = ndf.reset_index() 
ndf.to_csv('Monthly Sales.csv',index=False)
```

## Quarter Sale

```python
df['Quarter_usa'] = df['Order Date'].dt.quarter   
df
```

```python
# 2 Apr - june  1 
# 3 july - sep  2 
# 4 oct - dec   3 
# 1 jan-mar     4
```

```python
def q(quarter):
    if(quarter==2):
        return 1
    elif(quarter==3):
        return 2 
    elif(quarter==4):
        return 3
    else:
        return 4
```

```python
df['quarter'] = df['Quarter_usa'].apply(q)  

df
```

```python
df = df.drop(columns='Quarter_usa')
df
```

```python
ndf = df.groupby('quarter').agg(
    revenue = ('Amount','sum'),
    avg_sale = ('Amount','mean'),
    quantity = ('Quantity Ordered','sum')
)

ndf=ndf.reset_index()  
ndf
```

```python
plt.figure(figsize=(10,5))
ndf['quarter'] = ndf['quarter'].astype(str)
plt.plot(ndf['quarter'],ndf['revenue'],marker='o',label='Flipkart Sales',color='Skyblue')
plt.title('Flipkart Sales')
plt.grid() 
plt.legend()
plt.xlabel('Quarter')
plt.ylabel('Sale')
plt.xticks(rotation=45)
plt.show()
```
